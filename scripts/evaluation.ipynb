```python
import os 
import pandas as pd
import matplotlib.pylab as plt
# plt.style.use("seabor-v0_8")
```

```python
# Parse holdout set

# Open the file for reading
dir_path = "/Users/mmarzi/MLprojects/cv4e/evaluation/evaluation_data/pred_resol=1920_epochs=70_iou=0.2_holdout/labels"
file_names = [os.path.join(dir_path, f) for f in os.listdir(dir_path)]

df_holdout = pd.DataFrame(columns=["class", "surface", "confidence", "image_name", "img_inst_id"])

for f_name in file_names:

    with open(f_name, "r") as file:
        # Read all lines from the file
        lines = file.readlines()
        
        # Process each line
        line_data = []

        for count, inst in enumerate(lines):
            # Split the lines into individual values using space as the seperator
            instance = inst.strip().split()
            
            img_class = instance[0]
            surface = len(instance[1:-1])/2
            conf = instance[-1]
            img_name = os.path.basename(f_name)
            
            # Add to df
            df_holdout.loc[len(df_holdout)] = [img_class, surface, conf, img_name, count]

df_holdout
```

```python
# Parse test set

# Open the file for reading
dir_path = "/Users/mmarzi/MLprojects/cv4e/evaluation/evaluation_data/pred_resol=1920_epochs=70_iou=0.2_test/labels"
file_names = [os.path.join(dir_path, f) for f in os.listdir(dir_path)]

df_test = pd.DataFrame(columns=["class", "surface", "confidence", "image_name", "img_inst_id"])

for f_name in file_names:

    with open(f_name, "r") as file:
        # Read all lines from the file
        lines = file.readlines()
        
        # Process each line
        line_data = []

        for count, inst in enumerate(lines):
            # Split the lines into individual values using space as the seperator
            instance = inst.strip().split()
            
            img_class = instance[0]
            surface = len(instance[1:-1])/2
            conf = instance[-1]
            img_name = os.path.basename(f_name)
            
            # Add to df
            df_test.loc[len(df_test)] = [img_class, surface, conf, img_name, count]

df_test
```

```python
# Plot the test data set

df = df_test

# Aggregate surface
surface_counts = df['surface'].value_counts()

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 50, 50))

# Plot histogram using hist function
plt.figure(figsize=(10, 6))
plt.hist(df['surface'], bins=bins, edgecolor='black', alpha=0.7)
plt.xlabel('Surface (px)')
plt.ylabel('Frequency')
# plt.yscale("log")
plt.title('Distribution of phytoplankton surface (test set)')
plt.grid(True, which="both", ls="dashed")
plt.show()
```

```python
# Plot the test data set per class
df = df_test

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 50, 50))

# Create two subplots side by side
fig, ax = plt.subplots(1, 2, figsize=(15, 6), sharey=True) 

classes = ["0", "1"]

for i, cls in enumerate(classes):
    subset = df[df['class'] == cls]
    ax[i].hist(subset['surface'], bins=bins, edgecolor='black', alpha=0.7)
    ax[i].set_title(f'Class {cls}')
    ax[i].set_xlabel('Surface (px)')
    # ax[i].set_yscale('log')
    ax[i].grid(True, which="both", ls="dashed")

ax[0].set_ylabel('Frequency')

# plt.title('Distribution of phytoplankton surface (test set)')
plt.tight_layout()
plt.show()
```

```python
# Plot the holdout data set

df = df_holdout

# Aggregate surface
surface_counts = df['surface'].value_counts()

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 50, 50))

# Plot histogram using hist function
plt.figure(figsize=(10, 6))
plt.hist(df['surface'], bins=bins, edgecolor='black', alpha=0.7)
plt.xlabel('Surface (px)')
plt.ylabel('Frequency (log)')
plt.yscale("linear")
plt.title('Distribution of phytoplankton surface (holdout set)')
plt.grid(True, which="both", ls="dashed")
plt.show()
```
