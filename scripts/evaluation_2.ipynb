```python
import os 
import pandas as pd
import matplotlib.pylab as plt
```

```python
# Predicted test set df
df_test = pd.read_csv('/Users/mmarzi/MLprojects/cv4e/scripts/pred_test_annotations.csv')
df_test
```

```python
# Predicted holdout set df
df_holdout = pd.read_csv("/Users/mmarzi/MLprojects/cv4e/scripts/pred_holdout_annotations.csv")
df_holdout
```

```python
# open  Groun trouth annotatios
csv_path = "/Users/mmarzi/MLprojects/cv4e/scripts/GT_annotations.csv"

# Load the CSV into a dataframe
df_test_gt = pd.read_csv(csv_path)

# Mapping the values in the 'class' column
class_mapping = {
    "Pseudo-nitzschia-delicatissima": "0",
    "Pseudo-nitzschia-calliantha": "1"
 }
df_test_gt['class'] = df_test_gt['class'].map(class_mapping)

# # Convert for lower resolution
# ratio = (1920*1440) / (2584*1936) 
# df_test_gt["surface"] = df_test_gt["surface"] * ratio

# Display the first few rows of the dataframe
# Filter by class and sort by surface

df_test_gt[df_test_gt["class"] == "1"].sort_values(by="surface")
```

```python
# Plot the test data set

df = df_test

# Aggregate surface
surface_counts = df['surface'].value_counts()

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 1000, 1000))

# Plot histogram using hist function
plt.figure(figsize=(10, 6))
plt.hist(df['surface'], bins=bins, edgecolor='black', alpha=0.7)
plt.xlabel('Surface (px)')
plt.ylabel('Frequency')
# plt.yscale("log")
plt.title('Distribution of phytoplankton surface (test set)')
plt.grid(True, which="both", ls="dashed")
plt.show()
```

```python
# Plot the test data set per class
df = df_test

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 50, 50))

# Create two subplots side by side
fig, ax = plt.subplots(1, 2, figsize=(15, 6), sharey=True) 

classes = ["0", "1"]

for i, cls in enumerate(classes):
    subset = df[df['class'] == cls]
    ax[i].hist(subset['surface'], bins=bins, edgecolor='black', alpha=0.7)
    ax[i].set_title(f'Class {cls}')
    ax[i].set_xlabel('Surface (px)')
    # ax[i].set_yscale('log')
    ax[i].grid(True, which="both", ls="dashed")

ax[0].set_ylabel('Frequency')

# plt.title('Distribution of phytoplankton surface (test set)')
plt.tight_layout()
plt.show()
```

```python
# Plot the holdout data set

df = df_holdout

# Aggregate surface
surface_counts = df['surface'].value_counts()

# Create bins of 50 width
bins = list(range(0, int(df['surface'].max()) + 1000, 1000))

# Plot histogram using hist function
plt.figure(figsize=(10, 6))
plt.hist(df['surface'], bins=bins, edgecolor='black', alpha=0.7)
plt.xlabel('Surface (px)')
plt.ylabel('Frequency (log)')
plt.yscale("linear")
plt.title('Distribution of phytoplankton surface (holdout set)')
plt.grid(True, which="both", ls="dashed")
plt.show()
```
